```python
import ultralytics
print(ultralytics.__version__)
```

```python
# Training the YOLO v11 Model
from ultralytics import YOLO
import torch

# Path to your dataset YAML file (person-only subset of VisDrone)
DATA = r"VisDronePerson.yaml"

# Pick GPU if available, else fallback to CPU
device = 0 if torch.cuda.is_available() else "cpu"

# Load YOLOv11 pretrained weights (nano version for speed)
model = YOLO("yolo11n.pt")

# Train the model
model.train(
    data=DATA,
    epochs=100,               # number of training epochs
    imgsz=960,               # image size (can try 640/960/1280 depending on GPU)
    batch=8,                 # batch size (adjust based on GPU VRAM)
    device=device,           # 0 = first GPU, "cpu" = CPU
    workers=8,               # number of dataloader workers
    name="visdrone_person_v11n",  # experiment name
    patience=30              # early stopping if no improvement
)
```

```python
# Evaluate the Model
from ultralytics import YOLO

# Load best trained weights
model = YOLO("runs/detect/visdrone_person_v11n/weights/best.pt")

# Evaluate on validation set
metrics = model.val()
print(metrics)
```

```python
# Load best trained weights
model = YOLO("runs/detect/visdrone_person_v11n/weights/best.pt")

# Evaluate on testing set
metrics_test = model.val(data="VisDronePerson.yaml", split="test")

print(metrics_test)
```

```python
# Run Inference on Test Images
from ultralytics import YOLO

# Load trained model
model = YOLO("runs/detect/visdrone_person_v11n/weights/best.pt")

# Run prediction on a single image
results = model.predict(
    source="VisDronePerson_subset/images/test/9999938_00000_d_0000032.jpg",
    save=True,     # save predictions with bounding boxes
    conf=0.25,     # confidence threshold
    imgsz=960
)

# Loop through results programmatically
for r in results:
    print(r.boxes)   # bounding box data
```

```python
# Run prediction on all test images
model.predict(
    source="VisDronePerson_subset/images/test",
    save=True,
    conf=0.25,
    imgsz=960
)
```

```python
# Or run on a video
model.predict(
    source="drone_video.mp4",
    save=True,
    conf=0.25
)
```

```python
# Export Model for Deployment
model.export(format="onnx")   # export to ONNX
model.export(format="torchscript")
model.export(format="engine") # TensorRT (NVIDIA GPUs)
```
